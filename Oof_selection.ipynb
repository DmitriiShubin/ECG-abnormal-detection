```python
import pandas as pd
import numpy as np
```

```python
#load ECG dataset
df = pd.read_csv('data/data.csv',index_col=None,header=None)
df.head()
```

```python
oof = pd.DataFrame(np.zeros((100,129)))
for i in range(100):
    r = np.random.randint(0, high=df.shape[0])
    oof.loc[i] = df.iloc[r,:].values
    df = df.drop(r, axis=0)
    
oof.head()
```

```python
oof.to_csv('data/oof.csv')
df.to_csv('data/data_train.csv')
```
